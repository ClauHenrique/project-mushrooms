```python
# create test file
%%file test_data.py

import pytest
import wandb
import pandas as pd
import numpy as np

# initializing a wandb run to check preprocessing
run = wandb.init(entity="flamigos", project="Murshroom-Kmeans", job_type="check_data")

@pytest.fixture(scope="session")
def data():
    artifact = run.use_artifact("flamigos/Murshroom-Kmeans/clean_data_dataset:latest").file()
    df = pd.read_csv(artifact)

    return df


def test_data_length(data):
    # check if we have a required amount of data
    assert len(data) > 5000


def test_number_of_columns(data):
    # check number of columns in the dataset
    assert data.shape[1] == 23


def test_type(data):
    # check if all column values are integers
    assert np.issubdtype(data.values.dtype, np.integer) == True


def test_null_values(data):
  # test whether there is missing data in the data frame
  assert data.isnull().any().any() == False


def test_ranges(data):
  # test whether category values are within an expected range of values
  min = 0
  max = 12

  for col in data.columns:
    assert data[col].between(min, max).all()
```
